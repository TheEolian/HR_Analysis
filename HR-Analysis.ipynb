```python
df = pd.read_csv('HRDataset_v14.csv')
```

```python
df = df
```

```python
df.head()
```

```python
name_df = df['Employee_Name'].str.split(',', expand=True)
```

```python
df["Last_Name"]= name_df[0]
df["First_Name"]= name_df[1]
```

```python
df.head()
```

```python
gender = df['GenderID']
pay = df['Salary']
correlation = gender.corr(pay)
```
